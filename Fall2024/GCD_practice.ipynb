```python
import numpy as np
import time
import matplotlib.pyplot as plt

MAX = 1e6    #10**6

def Euclid_GCD(m, n): #Assuming m>=n>=0 with the exception of m=n=0, GCD(m=60, n=24) = 12
  if n==0:
    return m
  return Euclid_GCD(n, m%n)

def Bad_GCD(m, n):
  k = n
  while m%k or n%k : #eventually k will be decremented down to 12, which divides both m and n.
    k = k-1
  return k

def plotRuntimes(selected_gcd_algorithm, colorcode, label):
  runtime = []
  input_size = []
  for i in range(100):
    n = np.random.randint(1, MAX)
    m = np.random.randint(n, MAX) #make sure m>=n

    init_time = time.time()         #tic
    g = selected_gcd_algorithm(m, n)
    elapsed = time.time() - init_time  #toc
    runtime.append(elapsed)
    input_size.append(np.log2(m) + np.log2(n))

  plt.scatter(input_size, runtime, c=colorcode, label=label)  # to use different colors and legends

if __name__ == "__main__": #good practice when writing scripts that could later be imported as modules
  # Plotting both GCD algorithms
  plotRuntimes(Euclid_GCD, 'r', 'Euclid GCD')
  plotRuntimes(Bad_GCD, 'b', 'Bad GCD')
  plt.xlabel('Input Size (log2)')
  plt.ylabel('Time (seconds)')
  plt.title('Runtime of GCD Algorithms')
  plt.legend()
  plt.show()
```
