```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Sample data generation
np.random.seed(0)
n_samples = 30
x = np.linspace(-3, 3, n_samples)
noise = np.random.normal(0, 0.5, n_samples)
y_true = ((x+noise) > 0).astype(int)

X = x.reshape(-1,1)

# Logistic Regression model
model = LogisticRegression()
model.fit(X, y_true)

# Predict probabilities
X_test = np.linspace(-4, 4, 500).reshape(-1, 1)
probabilities = model.predict_proba(X_test)[:, 1]

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, y_true, color='black', zorder=3, label='Data points')
plt.plot(X_test, probabilities, color='red', zorder=2, linewidth=3, label='Logistic Regression Curve')
plt.xlabel('Feature Value (X)')
plt.ylabel('Probability')
plt.title('Logistic Regression Fit')
plt.legend()
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

feature_id = 2

# Load Iris dataset
iris = load_iris()
X = iris.data[iris.target != 0, feature_id].reshape(-1,1)  # select only classes 1 and 2, and only the first feature
y_true = iris.target[iris.target != 0] - 1  # adjust target to be 0 and 1 instead of 1 and 2

# Logistic Regression model
model = LogisticRegression()
model.fit(X, y_true)

# Predict probabilities
X_test = np.linspace(3, 7, 300).reshape(-1, 1)  # Adjust range to cover possible values of the first feature
probabilities = model.predict_proba(X_test)[:, 1]

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, y_true, color='black', label='Data points')
plt.plot(X_test, probabilities, color='red', label='Logistic Regression Curve')
plt.xlabel(iris.feature_names[feature_id])
plt.ylabel('Probability of Class 2')
plt.title('Logistic Regression on Iris Dataset (Classes 1 and 2)')
plt.legend()
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```
